```python
import numpy as np
#cost calculation for single iteration

def calculate_cost(x, y, weights):
    data_size = len(x)
    cost = 0
    for x,y in zip(x, y):
        y_pred = np.dot(weights, np.array([1.0, x]))
        cost+= pow((y_pred - y),2)
    
    cost = cost / (2.0*data_size)
    
    return cost
#calculate_cost(df[0],df[1],np.array([1,1]))
```
